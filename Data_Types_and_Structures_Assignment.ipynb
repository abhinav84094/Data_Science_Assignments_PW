8.What is a hash table, and how does it relate to dictionaries in Python?

Hashtable is a data structure in java or other programming language that stores data in the form of key-value. In python dictionay also stores data in the form of key-value.

A hash table is a data structure that stores data in key-value pairs, where each key is processed through a hash function to determine its position (index) in memory.
This allows for very fast access, insertion, and deletion operations.

Relation to Python Dictionaries:
- A Python dictionary is implemented internally using a hash table.
- When we store data in a dictionary, Python:
  - Calculates a hash value for the key.
  - Uses that hash to determine where the key-value pair is stored in memory.
  - Retrieves values in O(1) average time complexity (very fast).
```
# This is formatted as code
# Dictionary (internally uses hash table)
student = {"name": "Abhinav", "course": "BCA"}

# Accessing a value
print(student["name"])  # Output: Abhinav
```

10. Explain why strings are immutable in Python.

A string in Python is an immutable sequence of characters, meaning once it’s created, it cannot be changed.

Reasons Why Strings Are Immutable:
- Memory Efficiency (String Interning): Python often reuses string objects with the same value to save memory.
If strings were mutable, changing one string could unintentionally affect others sharing the same memory reference.
- Hashing and Dictionary Keys: Strings are frequently used as keys in dictionaries.
For keys to remain reliable, they must be hashable, and immutability ensures their hash value never changes.
- Security and Data Integrity: Since strings are often used to store sensitive data (like file paths, URLs, etc.), immutability prevents accidental or malicious modifications.
- Thread Safety: Immutable objects can be shared across multiple threads without synchronization issues, ensuring safer parallel execution.

11.What advantages do dictionaries offer over lists for certain tasks?

A dictionary is a collection of key-value pairs, while a list is an ordered collection of elements accessed by index.

Advantages of Dictionaries over Lists:
- Fast Lookup by Key:
  - Accessing a value in a dictionary using a key is O(1) on average (thanks to hash tables).
  - While In a list, searching for a value requires O(n) time (linear search).

- Meaningful Keys:
  - Dictionaries allow us to use descriptive keys instead of numeric indexes.
  - This makes code more readable and maintainable.

- Automatic Handling of Unique Keys:
  - Dictionary keys are unique by default, which is useful for tasks like counting or mapping.

- Flexible Data Association:
  - A dictionary can store complex data using nested structures, mapping keys to lists, sets, or other dictionaries.

- Efficient for Dynamic Updates:
  - Adding or removing key-value pairs is faster and cleaner than updating a large list with index calculations.

```
# This is formatted as code
# List lookup (slower)
students = ["Alice", "Bob", "Charlie"]
print("Bob" in students)  # O(n)

# Dictionary lookup (faster)
student_dict = {"Alice": 85, "Bob": 92, "Charlie": 78}
print(student_dict["Bob"])  # O(1)

```

14. How does the “in” keyword work differently for lists and dictionaries?

Lists:
- in checks if a value exists anywhere in the list.
- It performs a linear search, which takes O(n) time in the worst case.
```
# This is formatted as code
# List example
li = [10, 20, 30]
print(20 in li)   # True
print(40 in li)   # False
```
Dictionaries:
- in checks if a key exists in the dictionary, not the value.
- Since dictionaries use hash tables, key lookup is O(1) on average (very fast).
```
# This is formatted as code
# Dictionary example
dic = {"name": "Abhinav", "age": 21}
print("name" in dic)   # True
print("Abhinav" in dic) # False (checks key, not value)
```
In lists, in searches values sequentially, while in dictionaries, in checks keys using efficient hashing.





15.Can you modify the elements of a tuple? Explain why or why not.
- No, we cannot modify the elements of a tuple directly in Python.

Reason:
- Tuples are immutable, meaning once a tuple is created, its size and content cannot be changed.
- This immutability ensures data integrity, safe usage as dictionary keys, and faster performance compared to lists.
```
# This is formatted as code
# Tuple example
t = (1, 2, 3)

# Trying to modify an element
# t[0] = 10   # TypeError: 'tuple' object does not support item assignment

# You can create a new tuple instead

```

Tuples cannot be modified to ensure immutability, data safety, and consistent behavior.

16. What is a nested dictionary, and give an example of its use case?

A nested dictionary is a dictionary that contains another dictionary (or dictionaries) as its value.
This allows us to represent complex, hierarchical data in a structured way.

```
# This is formatted as code
nested_dict = {
    "student1": {"name": "Abhinav", "age": 21, "course": "BCA"},
    "student2": {"name": "Aman", "age": 22, "course": "BSc"}
}

print(nested_dict["student1"]["name"])  # Output: Abhinav
print(nested_dict["student2"]["course"]) # Output: BSc

```
