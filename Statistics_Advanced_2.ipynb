 1: What is hypothesis testing in statistics?
 - Hypothesis testing is a method in statistics used to make a decision about a claim or assumption (called a hypothesis) about a population, using data from a sample.

- Key ideas involved

  - Null hypothesis (H₀): The default assumption. It usually says nothing has changed or there is no effect.

  - Alternative hypothesis (H₁ or Hₐ): What you want to test for. It says there is a change or an effect exists.

  - Sample data: Information collected from a small group to represent the population.

  - Test statistic: A value calculated from the sample data.

  - p-value: The probability of getting results as extreme as the sample, assuming the null hypothesis is true.

Decision rule:

- If the p-value is small → reject the null hypothesis

- If the p-value is large → do not reject the null hypothesis

Example

- Suppose a school claims the average test score is 70.

- H₀: The average score is 70

- H₁: The average score is not 70

- You test a sample of students. If the results are very unlikely to happen when the average is really 70, you reject H₀.

- Why hypothesis testing is useful

- It helps make data-based decisions

- It reduces guesswork

- It’s widely used in science, medicine, business, and social studies

2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- The null hypothesis and the alternative hypothesis are two opposing statements used in hypothesis testing.

Null Hypothesis (H₀)

- It is the default or starting assumption

- States that there is no effect, no difference, or no change

- Any observed difference is assumed to be due to random chance

- Example:
  - H₀: The average test score is 70.


Alternative Hypothesis (H₁ or Hₐ)

- It is the claim you are testing for

- States that there is an effect, difference, or change

- Accepted only if there is strong evidence against the null hypothesis

- Example:
  - H₁: The average test score is not 70.

 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
- The significance level, denoted by α (alpha), is a threshold used in hypothesis testing to decide whether to reject the null hypothesis.

- The significance level is the maximum probability of making a Type I error
(rejecting a true null hypothesis).

- Common values are 0.05, 0.01, and 0.10.

Role in decision-making

- After calculating the p-value, it is compared with α:

- If p-value ≤ α → reject the null hypothesis

- If p-value > α → do not reject the null hypothesis

Why it matters

- It controls how strict the test is:

- Smaller α → stricter test, less chance of false positives

- Larger α → easier to reject H₀, but higher risk of error

- Example

  - If α = 0.05, you are willing to accept a 5% risk of concluding that an effect exists when it actually does not.

The significance level sets the cutoff for deciding whether results are statistically significant, guiding whether the evidence is strong enough to reject the null hypothesis.

4: What are Type I and Type II errors? Give examples of each.
- Type I and Type II Errors (with Examples)

- In hypothesis testing, decisions are made based on sample data. Sometimes, these decisions can be wrong. Such mistakes are called Type I and Type II errors.

1. Type I Error (α Error)
  - A Type I error occurs when the null hypothesis (H₀) is rejected even though it is actually true.

  - In simple words: False positive

  - Probability of committing a Type I error is denoted by α (alpha)

  - Example:
Suppose a court assumes that a person is innocent (null hypothesis).

  - H₀: The person is innocent

  - Decision: The court declares the person guilty, even though they are actually innocent

2. Type II Error (β Error)

  - A Type II error occurs when the null hypothesis (H₀) is not rejected even though it is false.

  - In simple words: False negative

  - Probability of committing a Type II error is denoted by β (beta)

  - Example:
  - Again, consider the court case:

  - H₀: The person is innocent

  - Reality: The person is actually guilty

  - Decision: The court declares the person not guilty

 5: What is the difference between a Z-test and a T-test? Explain when to use
each.
- In hypothesis testing, Z-test and T-test are statistical tests used to determine whether there is a significant difference between sample data and the population (or between two samples). The choice of test depends mainly on sample size and population standard deviation.

1. Z-test
- A Z-test is used when the population standard deviation (σ) is known and the sample size is large (n ≥ 30). It is based on the normal distribution.

- When to Use Z-Test:

  - Sample size ≥ 30

  - Population standard deviation is known

  - Data is normally distributed (or sample is large)

  - Used for testing means or proportions

  - Example:
  - A factory knows the population standard deviation of bulb life is 50 hours. - A sample of 100 bulbs is tested to check whether the average life is 1,000 hours.

2. T-test
- A T-test is used when the population standard deviation is unknown and the sample size is small (n < 30). It is based on the t-distribution.

- When to Use T-Test:

  - Sample size < 30

  - Population standard deviation is unknown

  - Data is approximately normally distributed

  - Common types: one-sample, two-sample, and paired T-test

  - Example:
  - A teacher measures the marks of 20 students to check whether the average mark differs from 60, but the population standard deviation is not known.


Use Z-test when the sample is large and population standard deviation is known.

Use T-test when the sample is small and population standard deviation is unknown.

```python
# 6: Write a Python program to generate a binomial distribution with n=10 and
# p=0.5, then plot its histogram.

# (Include your Python code and output in the code box below.)
# Hint: Generate random number using random function.

# Python program to generate a Binomial Distribution and plot histogram

import random
import matplotlib.pyplot as plt

n = 10
p = 0.5
samples = 1000

# Generate binomial distribution using random function
data = []
for _ in range(samples):
    successes = 0
    for _ in range(n):
        if random.random() < p:
            successes += 1
    data.append(successes)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

print("Sample output values:", data[:10])
```

```python
#  7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


import numpy as np
from scipy import stats

mu = 50 # population means assumed

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
# 8: Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)
sample_size = 100
population_mean = 50
population_std = 10

data = np.random.normal(population_mean, population_std, sample_size)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Calculate 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))

# Step 4: Plot the data
plt.hist(data, bins=15, edgecolor='black')
plt.axvline(sample_mean, linestyle='dashed', linewidth=2)
plt.title("Simulated Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#  9: Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores

# Sample dataset
data = [45, 48, 50, 52, 55, 49, 51, 53, 47, 54]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Z-scores:")
print(z_scores)

# Plot histogram of standardized data
plt.hist(z_scores, bins=8, edgecolor='black')
plt.axvline(0, linestyle='dashed', linewidth=2)
plt.title("Histogram of Z-Scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
