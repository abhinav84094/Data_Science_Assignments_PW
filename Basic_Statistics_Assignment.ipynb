1.  What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
- Descriptive statistics and inferential statistics are two main branches of statistics

Descriptive Statistics
  - Descriptive statistics summarize, organize, and describe the main features of a dataset. They do not make conclusions beyond the data collected.
  - It Describe what the data show
  - It Use numbers, tables, and graphs
  - Measures of central tendency: mean, median, mode
  - Measures of dispersion: range, variance, standard deviation
  - Graphs: bar charts, pie charts, histograms
  - Example :
    - Suppose a teacher records the exam scores of 30 students in a class.
      - Average (mean) score = 75
      - Highest score = 95
      - Lowest score = 50
      - Standard deviation = 10
  - These values describe the performance of that class only. No generalization is made to other classes.


Inferential Statistics
- Inferential statistics use sample data to make predictions or draw conclusions about a larger population.
- It Goes beyond the collected data
- It Estimates population characteristics
- It Test hypotheses
- Probability theory
- Confidence intervals
- Hypothesis testing (t-test, chi-square test, ANOVA)
- Regression analysi
- Example
  - From the same class, the teacher randomly selects 10 students and finds:
    - Mean score of sample = 78
    - Using inferential statistics, the teacher estimates that:
    - The average score of all students in the school is likely around 78, within a certain confidence level (e.g., 95%).
- This involves generalizing from a sample to a population.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
- Sampling is the process of selecting a subset (sample) from a larger group called the population, in order to study the characteristics of the whole population. Sampling is used because studying the entire population is often time-consuming, costly, or impractical.
- Example:
  - Instead of surveying all 10,000 voters in a city, a researcher selects 500 voters to represent the entire population.
- Random Sampling vs Stratified Sampling

i. Random Sampling
- In random sampling, every member of the population has an equal and independent chance of being selected.
- Selection is completely by chance
- No favoritism or bias
- Simple to use
- Example:
  - If a school has 1,000 students, and 100 students are selected using a lottery method or computer-generated random numbers, this is random sampling.
- It Reduces selection bias
- In this May not properly represent all subgroups of the population

ii. Stratified Sampling
- In stratified sampling, the population is first divided into subgroups (strata) based on certain characteristics (such as age, gender, income, or class), and then random samples are taken from each stratum.
- Ensures representation of all important subgroups
- More accurate than simple random sampling
- Example:
  - A company has employees as follows:
    - 60% men
    - 40% women
  - To select 100 employees, the researcher selects:
    - 60 men randomly
    - 40 women randomly
  - This is stratified sampling.
- In this Better representation of the population
- It is More complex and time-consuming


Random sampling: select individuals by chance from the whole population.

Stratified sampling: divide the population into groups and sample from each group.

3.  Define mean, median, and mode. Explain why these measures of central
tendency are important.
- Measures of central tendency are statistical values that describe the center or typical value of a dataset. The three main measures are mean, median, and mode.

i. Mean
- The mean is the arithmetic average of a set of values.
- Example
  - For the data: 2, 4, 6, 8, 10
  - Mean = (2+4+6+8+10)/5=6
  - Mean = 6

ii. Median
- The median is the middle value of a dataset when the data are arranged in ascending or descending order.
- If the number of observations is odd, the median is the middle value.
- If the number of observations is even, the median is the average of the two middle values.
- Example
  - For the data: 3, 5, 7, 9, 11
  - Median = 7

  - For the data: 2, 4, 6, 8
  - Median = (4+6)/2=5

iii. Mode
- The mode is the value that occurs most frequently in a dataset.
- Example
  - For the data: 1, 2, 2, 3, 4
  - Mode = 2
- A dataset may have:
  - One mode (unimodal)
  - Two modes (bimodal)
  - More than two modes (multimodal)
  - No mode (all values occur once)
- Importance of Measures of Central Tendency
  - Summarize large data : They provide a single value that represents an entire dataset.
  - Easy comparison : They help compare different datasets quickly (e.g., average marks of two classes).
  - Decision-making : Used in business, education, economics, and research for planning and policy decisions.
  - Understanding data distribution : Comparing mean, median, and mode helps identify skewness and outliers.
  - Basis for further analysis : They form the foundation for advanced statistical analysis.

4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
- Skewness and kurtosis are measures that describe the shape of a data distribution, especially how it deviates from a normal distribution.

i. Skewness
- Skewness measures the degree of asymmetry of a distribution around its mean.
- Types of Skewness
  - Positive Skewness (Right-skewed) : The tail on the right side of the distribution is longer.
  - Most data values are concentrated on the left.
- Mean>Median>Mode

ii. Negative Skewness (Left-skewed) : The tail on the left side is longer.
- Most data values are concentrated on the right.
- Mean<Median<Mode

Zero Skewness : Distribution is symmetrical (normal distribution).
- Mean = Median = Mode

- Example of Positive Skew
  - Income distribution is usually positively skewed:
  - Most people earn low to moderate incomes.
  - A few people earn very high incomes, stretching the right tail.

ii. Kurtosis : Kurtosis measures the peakedness or flatness of a distribution compared to a normal distribution.
- Types of Kurtosis
  - Leptokurtic
  - Sharp peak
  - Heavy tails
  - More extreme values (outliers)
  - Mesokurtic
  - Normal peak
  - Similar to normal distribution
  - Platykurtic
  - Flat peak
  - Light tails
  - Fewer extreme values
- Example
  - Stock returns often show leptokurtic distributions due to extreme price changes.

What Does Positive Skew Imply About the Data?

A positive skew implies that:
- Most observations are smaller values
- A few very large values pull the mean to the right
- The mean is greater than the median
- There is a long right tail


Skewness: tells us if data lean left or right

Kurtosis: tells us how sharp or flat the peak is

Positive skew: most values are low, with some very high values

```python
# 6.Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

import numpy as np

cov_matrix = np.cov(list_x, list_y)

corr_matrix = np.corrcoef(list_x, list_y)

print("Covariance matrix:\n", cov_matrix)
print()
print("Correlation coefficient matrix:\n", corr_matrix)
```

```python
# 7.Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import numpy as np
import matplotlib.pyplot as plt

arr = np.array(data)

Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = arr[(arr < lower_bound) | (arr > upper_bound)]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)

plt.boxplot(arr, vert=False)
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()


```

```python
#8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# ● Explain how you would use covariance and correlation to explore this relationship.

''' How to use covariance and correlation to explore the relationship

As a data analyst, we can use covariance and correlation to understand how advertising spend and daily sales move together.

Covariance : Covariance indicates the direction of the relationship.
If covariance is:
  Positive → advertising spend and sales increase together
  Negative → one increases while the other decreases
  Zero → no linear relationship
Limitation: its value depends on the scale of data, so it is hard to interpret on its own.

Correlation : Correlation coefficient standardizes covariance.
  It measures both strength and direction of the relationship.
  Values range from –1 to +1:
    +1 → perfect positive relationship
    0 → no linear relationship
    –1 → perfect negative relationship
Correlation is preferred for decision-making because it is easy to interpret.

In this case, a strong positive correlation would suggest that increasing advertising spend is associated with higher daily sales.
'''

# ● Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]

daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

correlation_matrix = np.corrcoef(advertising_spend, daily_sales)
correlation = correlation_matrix[0, 1]

print("Correlation :\n", correlation)

# Interpretation of the Result :
# The correlation coefficient ≈ 0.99

# This indicates a very strong positive linear relationship between advertising spend and daily sales.

# As advertising spend increases, daily sales tend to increase significantly.
# The analysis suggests that higher advertising spend is strongly associated with higher daily sales, supporting the marketing team’s strategy to invest more in advertising (while noting that correlation does not prove causation).
```
