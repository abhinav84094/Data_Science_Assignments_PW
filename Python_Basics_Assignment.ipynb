Q7.  What is the difference between “==” and “is” operators in Python?
- Ans: Both == and is are comparison operators in Python, but they are used for different purposes:
- == (Equality Operator):
It compares the values of two objects to check if they are equal.
Example:
```
# This is formatted as code
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)
```
- is (Identity Operator):
It checks whether both variables refer to the same object in memory.
Example:
```
a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)
```
The == operator checks whether the values of two objects are the same.

The is operator checks whether both variables refer to the same memory address (object).

Q13. What is the difference between for and while loops?

````python
# Q6.Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_values(a_list, a_number):
    a_list.append(100)   # modifies list (mutable)
    a_number += 100      # creates a new integer (immutable)
    print("Inside function:")
    print("List:", a_list)
    print("Number:", a_number)

my_list = [1, 2, 3]
my_number = 10

modify_values(my_list, my_number)

print("\nOutside function:")
print("List:", my_list)
print("Number:", my_number)


# Mutable objects (like list, dict, set) → changes inside a function affect the original object.
# Immutable objects (like int, str, tuple) → changes inside a function do not affect the original.
````

````python
# Q8.Write a program to demonstrate the use of logical operators.
st1 = True
st2 = False

logicalAnd = st1 and st2
logicalOr = st1 or st2
print(logicalAnd)
print(logicalOr)


# Demonstration of logical operators

x = 10
y = 5
z = 15

print("x =", x, ", y =", y, ", z =", z)

# Using 'and' operator
print("\nUsing 'and' operator:")
print("x > y and x < z:", x > y and x < z)  # True and True → True
print("x > y and x > z:", x > y and x > z)  # True and False → False

# Using 'or' operator
print("\nUsing 'or' operator:")
print("x < y or x < z:", x < y or x < z)    # False or True → True
print("x < y or x > z:", x < y or x > z)    # False or False → False

# Using 'not' operator
print("\nUsing 'not' operator:")
print("not(x > y):", not(x > y))            # not(True) → False
print("not(y > z):", not(y > z))            # not(False) → True
````

````python
# Q14. Write a program to reverse a string using a while loop.
str1 = "abcdef"
revstr = ""
strlen = len(str1)
i = strlen-1
while i>=0:
  revstr += str1[i]
  i-=1

print("original string :", str1)
print("reversed string :", revstr)
````
