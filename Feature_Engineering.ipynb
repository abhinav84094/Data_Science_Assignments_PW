2. What is correlation?
What does negative correlation mean?
- Correlation is a statistical measure that describes the degree and direction of relationship between two variables.
It tells us how one variable changes in relation to another.
- Example:
    - Height and weight → usually positively correlated
    - Price and demand → often negatively correlated
- Correlation is measured using a value called the correlation coefficient (r):
  - r = +1 → Perfect positive correlation
  - r = 0 → No correlation
  - r = −1 → Perfect negative correlation

- A negative correlation means that when one variable increases, the other decreases, and vice versa.
- Example of Negative Correlation
  - As price increases, demand decreases
  - As speed increases, time taken decreases
  - As exercise increases, body fat decreases

6. How do we handle categorical variables in Machine Learning? What are the common techniques?
- Categorical variables cannot be directly used by most machine learning algorithms because they work with numerical data. Therefore, categorical variables must be converted into numerical form using encoding techniques.
- Label Encoding : Assigns a unique integer to each category.
- Ordinal Encoding : Assigns numbers based on ranking/order.
- Frequency (Count) Encoding : Replaces categories with their frequency or count.
- Target Encoding (Mean Encoding) : Replaces each category with the mean of the target variable.

11. Why do we have to perform EDA before fitting a model to the data?
- Exploratory Data Analysis (EDA) is performed before model fitting to understand the dataset, identify problems, and make informed preprocessing and modeling decisions.
Without EDA, a model may be built on poor-quality or misleading data, leading to inaccurate results.
- Key Reasons for Performing EDA
  - Understand the Data : Helps understand the structure, size, and nature of the data
  - Detect Missing Values : Identifies missing or null values
  - Identify Outliers : Outliers can distort model learning
  - Understand Data Distribution : Checks whether data is normal, skewed, or imbalanced
  - Identify Relationships Between Variables : Correlation analysis reveals relationships between features and target
  - Guide Feature Engineering : Helps create new meaningful features

 EDA is performed before model fitting to understand the data, detect issues like missing values and outliers, and make informed preprocessing and modeling decisions.

13. What does negative correlation mean?

````python
# 14. How can you find correlation between variables in Python?
import pandas as pd

data = {
    'Age': [20, 25, 32, 35, 40],
    'Salary': [20000, 30000, 40000, 50000, 60000]
}

df = pd.DataFrame(data)

correlation = df['Age'].corr(df['Salary'])
print(correlation)
df.corr()
````

15. What is causation? Explain difference between correlation and causation with an example.
- Causation refers to a relationship where one variable directly causes a change in another variable.
If A causes B, then a change in A produces a change in B.

| Basis      | Correlation                                    | Causation                                   |
| ---------- | ---------------------------------------------- | ------------------------------------------- |
| Meaning    | Measures the **association** between variables | Shows a **cause-and-effect** relationship   |
| Direction  | Variables may move together                    | One variable **directly affects** the other |
| Proof      | Does **not prove** cause                       | Requires logical or experimental proof      |
| Dependency | No dependency implied                          | Cause precedes effect                       |
| Example    | Ice-cream sales and drowning                   | Smoking causes lung disease                 |


16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- An optimizer is an algorithm used in machine learning and deep learning to update model parameters (weights and biases) in order to minimize the loss function.
It controls how fast and in which direction a model learns during training.
- An optimizer is a technique that adjusts a model’s parameters to minimize the loss function and improve model performance.

Types of Optimizers
- Gradient Descent (GD)
  - Updates parameters using the entire dataset
  - Moves in the direction of negative gradient
  - Example - Used in Linear Regression for small datasets
- Stochastic Gradient Descent (SGD)
  - Updates parameters after every single data point
  - Faster but more noisy than GD
  - Example - Used in online learning and large datasets.
- Mini-Batch Gradient Descent
  - Uses a small batch of data (e.g., 32, 64)
  - Combination of GD and SGD
  - Example - Most deep learning models use this.
- Momentum Optimizer
  - Uses past gradients to speed up learning
  - Reduces oscillations
  - Example - Training deep neural networks
- AdaGrad (Adaptive Gradient)
  - Learning rate is adapted for each parameter
  - Good for sparse data
  - Example - Used in text and NLP models
- RMSprop
  - Fixes AdaGrad’s learning-rate problem
  - Uses moving average of squared gradients
  - Example - Used in RNNs
- Adam (Adaptive Moment Estimation)
  - Combines Momentum + RMSprop
  - Adaptive learning rate + fast convergence
  - Example - Used in CNNs, Transformers, Deep Learning models

21. What is feature scaling? How does it help in Machine Learning?
- Feature scaling is a data preprocessing technique used in Machine Learning to rescale numerical features to a common range or scale so that no single feature dominates the learning process.
- Feature scaling is the process of transforming features so that they have comparable ranges or distributions.
- In real datasets, features often have different units and ranges:
  - Age: 18–60
  - Salary: 10,000–1,000,000
  - Without scaling, models may give more importance to large-valued features.

How Does Feature Scaling Help in Machine Learning?
- Improves Model Performance
  - Prevents dominance of large-scale features
  - Leads to better accuracy
- Faster Convergence
  - Gradient-based algorithms converge faster
  - Reduces training time
- Required for Distance-Based Algorithms
  - K-Nearest Neighbors (KNN)
  - K-Means Clustering
  - Support Vector Machines (SVM)
- Improves Numerical Stability
  - Prevents overflow/underflow issues
  - Helps in stable optimization
- Fair Feature Contribution
  - All features contribute equally
  - Important for linear models

Feature scaling is the process of bringing all features to a common scale to improve model performance and training efficiency.

````python
# 22. How do we perform scaling in Python?

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

data = {
    'Age': [18, 22, 25, 30, 35, 40],
    'Salary': [15000, 22000, 28000, 40000, 55000, 70000]
}

df = pd.DataFrame(data)

X = df[['Age', 'Salary']]

X_train, X_test = train_test_split(
    X, test_size=0.2, random_state=42
)


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Standard Scaled Training Data:")
print(X_train_scaled)

print("\nStandard Scaled Testing Data:")
print(X_test_scaled)



# Feature scaling in Python is performed using StandardScaler or MinMaxScaler from the scikit-learn to bring all features to a common scale.
````
