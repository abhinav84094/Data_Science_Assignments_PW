2.  How does broadcasting work in NumPy?
- NumPy broadcasting is a rule that allows NumPy to perform operations on arrays of different shapes without explicitly resizing them.
It automatically “stretches” smaller arrays so they can match the shape of larger ones.

```
a = np.array([1, 2, 3])
b = 5

a + b  # b becomes [5,5,5]
```

4. Explain the use of the groupby() method in Pandas.
- The groupby() method in Pandas is used to split data into groups based on some column(s), then apply some operation (like sum, mean, count), and finally combine the results.

```
import pandas as pd

data = {
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"],
    "Sales": [200, 150, 300, 100, 250],
    "Category": ["A", "B", "A", "B", "A"]
}

df = pd.DataFrame(data)
df.groupby("City")["Sales"].sum()
df.groupby("Category")["Sales"].mean()
```

- groupby() lets you analyze your data by categories.
It groups rows and then applies calculations on each group.

6.  What are the differences between NumPy arrays and Python lists?
- Speed (NumPy is much faster)
  - NumPy arrays are stored in contiguous memory and use optimized C code → operations are faster.
  - Example:
    - Mathematical operations on NumPy arrays can be 10x–100x faster than Python lists.
  - Python lists = slower because they store objects.

- Memory efficiency (NumPy uses less memory)
  - NumPy arrays store homogeneous data (same type), so memory usage is compact.
  - Python lists store pointers to objects, so they take more space.

- Data type restrictions
  - NumPy array: all elements must be of the same type
  - Python list: can store mixed data types

- Functionality (NumPy has more features)
  - NumPy provides:
    - Linear algebra
    - Statistics
    - Random numbers
    - Broadcasting
  - Python lists don’t support these operations.

- NumPy arrays are faster, more efficient, and built for numerical & scientific computing.
Python lists are general-purpose containers, slower, and not suitable for heavy math operations.

8. What does the term “vectorized operation” mean in NumPy?
- In NumPy, a vectorized operation means performing a mathematical operation on entire arrays at once, without using explicit Python loops (for loops).

```
# Normal Python
lst = [1, 2, 3]
result = [x * 2 for x in lst]

# Numpy vectorized
import numpy as np

arr = np.array([1, 2, 3])
result = arr * 2     # No loop needed
```

15. How does NumPy handle multidimensional arrays?
- NumPy handles multidimensional arrays using its core object called ndarray, which can have any number of dimensions (1D, 2D, 3D, …, nD). These arrays allow you to store and perform operations on homogeneous data efficiently.

```
import numpy as np

# 1D array
arr1 = np.array([1, 2, 3])
print(arr1.shape)  # (3,)

# 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2.shape)  # (2, 3)

# 3D array
arr3 = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print(arr3.shape)  # (2, 2, 2)
```

20. What is the role of a pivot table in Pandas?
- A pivot table in Pandas is a powerful tool that allows you to summarize, aggregate, and reorganize data in a tabular form. It is similar to Excel pivot tables and is widely used for data analysis and reporting.
- Pivot tables allow you to group data by one or more keys and calculate aggregate statistics such as sum, mean, count, min, max.

```
import pandas as pd

data = {
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Chennai'],
    'Category': ['A', 'A', 'B', 'B', 'A'],
    'Sales': [200, 150, 300, 250, 100]
}

df = pd.DataFrame(data)

pivot = df.pivot_table(values='Sales', index='City', columns='Category', aggfunc='sum')
print(pivot)
```




21. Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy’s array slicing is faster than Python’s list slicing because of how NumPy stores data and handles operations internally.
- Contiguous Memory Storage
  - NumPy arrays are stored in contiguous blocks of memory (like C arrays).
  - Python lists store references to objects, scattered in memory.
- Optimized C Implementation
  - NumPy is implemented in C, so slicing and operations are compiled and optimized.
  - Python lists are purely interpreted, which is slower for large datasets.

````python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Age': [25, 30, 35, 40, 28, 33],
    'Salary': [50000, 60000, 70000, 80000, 55000, 65000],
    'Experience': [2, 5, 8, 10, 3, 7]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
````
