 1: What is a random variable in probability theory?
 - In probability theory, a random variable is a function that assigns a numerical value to each outcome of a random experiment.
 - When an experiment has uncertain outcomes, we use a random variable to convert those outcomes into numbers so we can analyze them mathematically.
 - Example 1: Tossing a Coin

Let

X = number of heads obtained

Possible outcomes: {H, T}
Random variable values:

X = 1 if Head

X = 0 if Tail                                               

 - A random variable is a function that maps the outcomes of a random experiment to real numbers.

 2: What are the types of random variables?
 - In probability theory, there are two main types of random variables:
 - 1. Discrete Random Variable
    - Takes countable values
    - Values are usually integers
    - Probabilities are given by a Probability Mass Function (PMF)
  - Examples:
    - Number of heads in coin tosses
    - Number of students in a class
    - Number on a dice (1‚Äì6)

  - 2. Continuous Random Variable
    - Takes uncountable values within an interval
    - Probabilities are given by a Probability Density Function (PDF)
    - Probability at an exact point is zero
  - Examples:
    - Height of a person
    - Time taken to complete a task
    - Temperature

3: Explain the difference between discrete and continuous distributions.
- Discrete Distribution
  - Deals with discrete random variables
  - Possible values are countable
  - Uses Probability Mass Function (PMF)
  - Probability of a value can be non-zero
  - Probabilities are found by summing
  - Example:
    - Number of heads in 3 coin tosses
- Continuous Distribution
  - Deals with continuous random variables
  - Possible values are uncountable (within an interval)
  - Uses Probability Density Function (PDF)
  - Probability at an exact point is zero
  - Probabilities are found by integration
  - Example:
    - Height of students in a class

6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that:

The distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the shape of the population distribution, provided the samples are independent and identically distributed.

 7: What is the significance of confidence intervals in statistical analysis?
 - Significance of Confidence Intervals in Statistical Analysis

A confidence interval (CI) is a range of values used to estimate a population parameter (such as mean or proportion) based on sample data.

- Why Confidence Intervals Are Important
  - Measure of Uncertainty -
Shows how reliable an estimate is instead of giving a single value.
  - Population Parameter Estimation -
Provides a plausible range where the true population value lies.
  - Better than Point Estimates -
More informative than a single number.
  - Used in Decision Making - Helps in hypothesis testing and comparisons.
  - Level of Confidence - Common confidence levels are 90%, 95%, and 99%.

8: What is the concept of expected value in a probability distribution?
- The expected value (mean) of a random variable is the long-run average value of the variable over many repetitions of an experiment.
- The expected value is the weighted average of all possible values of a random variable, where the weights are their probabilities.
- The expected value of a probability distribution is the long-term average outcome of a random variable.

```python
#  9: Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

num = np.random.normal(loc=50, scale=5, size=1000)
mean_val = num.mean()
std_val = num.std()
print("Mean = ", mean_val)
print("Standard Deviation = ", std_val)

plt.hist(num)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution (Mean = 50, Std = 5)")
plt.show()
```

```python
# 10: You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
# 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# ‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
# with a 95% confidence interval.
# ‚óè Write the Python code to compute the mean sales and its confidence interval.


daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

import numpy as np
from scipy import stats

# Sample size
n = len(daily_sales)

# Sample mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample std

# Standard Error
se = std_sales / np.sqrt(n)

# 95% confidence interval using z-score 1.96
z = 1.96
ci_lower = mean_sales - z * se
ci_upper = mean_sales + z * se

print(f"Mean Daily Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")



# np.mean(daily_sales) ‚Üí computes sample mean
# np.std(daily_sales, ddof=1) ‚Üí computes sample standard deviation
# SE = s / ‚àön ‚Üí standard error
# Multiply SE by z-score for 95% confidence
# CI gives a range estimate for the true average daily sales
```
